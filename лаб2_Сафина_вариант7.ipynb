# 2.2.8

```
import matplotlib.pyplot as plt
langs = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
langs_pos = [i for i, _ in enumerate(langs)]
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language\n Worldwide, Oct 2017 compared to a year ago")
plt.xticks(langs_pos, langs)
width = [0.1,0.2,0.5,1.1,0.2,0.3]
y_pos = [0,0.8,1.5,3,5,6]
plt.bar(y_pos, popularity, width=width)
plt.xticks(y_pos, langs)
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.ylim([0,25])
plt.xlim([0,6])
plt.show()
```

# 2.3.3

```
import matplotlib.pyplot as plt
langs = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
fig = plt.figure()
explode = (0.1, 0, 0, 0, 0, 0.1)  
plt.pie(popularity, explode=explode, labels=langs, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("PopularitY of Programming Language\n Worldwide, Oct 2017 compared to a year ago", bbox={'facecolor':'0.8', 'pad':5})
plt.show()
```

```
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
countries = ['US','India','UK','Japan','Korea']
film_ammount = [3297,990,723,287,212]
ax.pie(film_ammount, labels = countries, autopct = '%1.2f%%',shadow=True,startangle=140)
plt.title("Films produced in different \n countries", bbox={'facecolor':'0.9', 'pad':6})
plt.show()
```

```
import matplotlib.pyplot as plt
countries = ['US','India','UK','Japan','Korea']
film_ammount = [3297,990,723,287,212]
countries_pos = [i for i, _ in enumerate(countries)]
plt.xlabel("Countries")
plt.ylabel("Ammount of produced films \n in each country")
plt.title("Produced films in the world")
plt.xticks(countries_pos, countries)
width = [0.1,0.2,0.5,1.1,0.2]
y_pos = [0,0.8,1.5,3,5]
plt.bar(y_pos, film_ammount, width=width)
plt.xticks(y_pos, countries)
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='green')
plt.ylim([0,3500])
plt.xlim([0,6])
plt.show()
```
